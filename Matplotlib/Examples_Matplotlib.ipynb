```python
import matplotlib.pyplot as plt
import numpy as np

# %matplotlib inline  # Uncomment if using Jupyter

# Sample data for demonstration
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
```

```python
# First example
plt.figure(figsize=(6, 4))  # Create a new figure with custom size
plt.plot(x, y, label='sin(x)', color='blue', linestyle='--', linewidth=2)
plt.title("Line Plot of Sine Wave")
plt.xlabel("X values")
plt.ylabel("sin(x)")
plt.legend()        # Show a legend
plt.grid(True)      # Add grid lines for better readability
plt.show()          # Display the plot
```

```python
# Generating some random data for demonstration
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, c='red', marker='o', alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
#  BAR PlOT
# ------------------------------------------------------------

categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='green')
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#  HISTOGRAM
# ------------------------------------------------------------
data = np.random.randn(1000)  # Normal distribution
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='purple', alpha=0.7, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# BOX PLOT
# ------------------------------------------------------------
# Creating multiple datasets to show box plots side by side
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(5, 2, 100)
data3 = np.random.normal(-3, 1.5, 100)

plt.figure(figsize=(6, 4))
plt.boxplot([data1, data2, data3], tick_labels=['Set 1', 'Set 2', 'Set 3'])
plt.title("Box Plot Example")
plt.ylabel("Value")
plt.show()
```

```python
# Pie chart
# ------------------------------------------------------------
slices = [40, 25, 20, 15]
labels = ['Slice A', 'Slice B', 'Slice C', 'Slice D']
explode = (0, 0.1, 0, 0)  # Offset the 2nd slice a bit

plt.figure(figsize=(6, 4))
plt.pie(slices, labels=labels, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

```python

# ------------------------------------------------------------
#TITLES, LABELS, LEGENDS, COLORS, AND LINE STYLES
# ------------------------------------------------------------
# Example with multiple lines, customizing styles
plt.figure(figsize=(6, 4))
plt.plot(x, np.sin(x), label='sin(x)', color='blue', linestyle='-')
plt.plot(x, np.cos(x), label='cos(x)', color='red', linestyle='--')

# Adding title, x/y labels, legend
plt.title("Multiple Lines with Custom Styles")
plt.xlabel("X Axis")
plt.ylabel("Function Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# ------------------------------------------------------------
# AXIS LIMITS AND TICK CUSTOMIZATION
# ------------------------------------------------------------
plt.figure(figsize=(6, 4))
plt.plot(x, y, color='red')

# Manually set x and y axis limits
plt.xlim(0, 2*np.pi)
plt.ylim(-2, 2)

plt.title("Axis Limits and Custom Tick Labels")
plt.xticks([0, np.pi, 2*np.pi], ['0', '', '2'])  # Custom X tick labels
plt.yticks([-1, 0, 1])
plt.grid(True)
plt.show()
```

```python
# ------------------------------------------------------------
# SUBPLOTS (SIDE BY SIDE)
# ------------------------------------------------------------
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# First subplot
axes[0].plot(x, np.sin(x), color='blue')
axes[0].set_title("Sine")

# Second subplot
axes[1].plot(x, np.cos(x), color='green')
axes[1].set_title("Cosine")

plt.suptitle("Subplots Example", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------------------
# MULTIPLE FIGURES
# ------------------------------------------------------------
# Figure 1
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='sin(x)')
plt.title("Figure 1")
plt.legend()
plt.show()

# Figure 2
plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, c='orange')
plt.title("Figure 2")
plt.show()
```

```python
# ------------------------------------------------------------
#  USING A BUILT-IN STYLE AND ANNOTATIONS
# ------------------------------------------------------------
plt.style.use('bmh')  # Switch to ggplot style

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='sin(x)')

# Annotate a specific point
x_point = np.pi
y_point = np.sin(x_point)
plt.scatter(x_point, y_point, color='black')
plt.annotate('Value at ', xy=(x_point, y_point),
             xytext=(x_point+0.5, y_point),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title("Styled Plot with Annotations")
plt.xlabel("X")
plt.ylabel("Y = sin(x)")
plt.legend()
plt.show()
```

```python
# ------------------------------------------------------------
# PRACTICAL EXAMPLE
# ------------------------------------------------------------
# Synthetic data
time = np.linspace(0, 10, 50)
values_line = 2 * time + 1
values_scatter = values_line + np.random.normal(0, 3, size=time.size)

plt.figure(figsize=(7, 5))
plt.plot(time, values_line, label='Ideal Trend', color='darkblue', linewidth=2)
plt.scatter(time, values_scatter, label='Observed Data', color='red', alpha=0.7)

plt.title("Line + Scatter Plot Example")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
